```python
#import libraries
import pandas as pd
import numpy as np
from dateutil.relativedelta import relativedelta
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
```

### Creating Table for Employee Groupings Analysis

```python
#Generate the intermediate df (without the age bracket and tenure)

#import file
headcount = 'Forvis Mazars Data Request\Headcount - Employee Dates and Status - 2021 to 2025 - Duplicates Dropped.xlsx'
headcount_df = pd.read_excel(headcount)

#config 
as_of_date = pd.Timestamp("2025-12-31")
year_value = 2025

df = headcount_df.copy()

#rename the columns
df = df.rename(columns={
    "Payroll Name": "employee_name",
    "Hire Date": "hire_date",
    "Termination Date": "termination_date",
    "Birth Year": "birth_date"
})

#convert dates
df["hire_date"] = pd.to_datetime(df["hire_date"], errors="coerce")
df["termination_date"] = pd.to_datetime(df["termination_date"], errors="coerce")
df["birth_date"] = pd.to_datetime(df["birth_date"], errors="coerce")

#filter original dataset
cutoff_date = pd.Timestamp("2025-01-01")
df["termination_date"] = pd.to_datetime(df["termination_date"], errors="coerce")
df = df.loc[df["termination_date"].isna() | (df["termination_date"] >= cutoff_date)].copy()
df.reset_index(drop=True, inplace=True)
#print("Terminations before 2025 present?:", (df["termination_date"] < cutoff_date).any())

#helper functions
def age_in_years(birth_date, as_of):
    if pd.isna(birth_date):
        return np.nan
    return relativedelta(as_of, birth_date).years

def tenure_years_months(start_date, end_date):
    if pd.isna(start_date) or pd.isna(end_date) or end_date < start_date:
        return None
    rd = relativedelta(end_date, start_date)
    return f"{rd.years} years {rd.months} months"

tenure_end = df["termination_date"].where(
    df["termination_date"].notna() & (df["termination_date"] <= as_of_date),
    as_of_date
)

#conversions
df["age"] = df["birth_date"].apply(lambda d: age_in_years(d, as_of_date)).astype("Int64")

df["tenure"] = [tenure_years_months(hire, end) for hire, end in zip(df["hire_date"], tenure_end)]

df["year"] = year_value

#turnover yes or no
df["turnover"] = np.where(df["termination_date"].notna() & (df["termination_date"] <= as_of_date), "Yes", "No")


#output
output_df = df[[
    "employee_name",
    "age",
    "tenure",
    "year",
    "turnover"
]].copy()

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
output_df.head(25)
```

```python
#add age bracket and tenure bracket columns to output_df
def get_age_bracket(birth_date):
    if pd.isna(birth_date):
        return None
    year = birth_date.year
    if 1928 <= year <= 1945:
        return 'Post War'
    elif 1946 <= year <= 1964:
        return 'Baby Boomers'
    elif 1965 <= year <= 1980:
        return 'Gen X'
    elif 1981 <= year <= 1996:
        return 'Millennials'
    elif 1997 <= year <= 2012:
        return 'Gen Z'
    else:
        return None

def get_tenure_bracket(tenure):
    if pd.isna(tenure) or tenure is None:
        return None
    try:
        years = int(str(tenure).split(' ')[0])
    except:
        return None
    if years < 2:
        return '<2 years'
    elif 2 <= years < 5:
        return '2-5 years'
    elif 5 <= years < 10:
        return '5-10 years'
    elif 10 <= years < 20:
        return '10-20 years'
    elif years >= 20:
        return '20+ years'
    else:
        return None

output_df['age_bracket'] = df['birth_date'].apply(get_age_bracket)
output_df['tenure_bracket'] = df['tenure'].apply(get_tenure_bracket)
output_df.head(25)
```

```python
#import and clean the termination report

#import file
termination_file = 'Forvis Mazars Data Request/termination report & payouts 2025.xlsx'
df_termination = pd.read_excel(termination_file)

import re
def clean_reason(reason):
    if pd.isnull(reason):
        return reason
    return re.sub(r'^[A-Za-z0-9]+\s*-\s*', '', str(reason)).strip()

#update the reason column
if 'REASON' in df_termination.columns:
    df_termination['REASON_CLEAN'] = df_termination['REASON'].apply(clean_reason)

    #mapping for cleaned responses to grouped
    def group_reason(reason):
        if pd.isnull(reason):
            return 'Terminated'
        r = reason.strip().lower()
        if r == 'resignation':
            return 'Resignation'
        elif r == 'retired' or r == 'retirement':
            return 'Retired'
        elif r == 'deceased':
            return 'Deceased'
        else:
            return 'Terminated'

    df_termination['REASON_GROUPED'] = df_termination['REASON_CLEAN'].apply(group_reason)
    reason_counts = df_termination['REASON_GROUPED'].value_counts(dropna=False)
else:
    print('REASON column not found in the file.')

df_termination.head(5)
```

```python
#merge REASON_GROUPED to Status

#match olumn types
headcount_df['Payroll Name'] = headcount_df['Payroll Name'].astype(str)
df_termination['NAME'] = df_termination['NAME'].astype(str)

#create mapping
reason_map = df_termination.set_index('NAME')['REASON_GROUPED'].to_dict()

#replace status when match
def update_status(row):
    if row['Payroll Name'] in reason_map:
        return reason_map[row['Payroll Name']]
    else:
        return row['Status'] if 'Status' in row else None

headcount_df['Status'] = headcount_df.apply(update_status, axis=1)
```

```python
#add status to output_df without duplicates
if 'Status' in output_df.columns:
    output_df = output_df.drop(columns=['Status'])
output_df = output_df.merge(headcount_df[['Payroll Name', 'Status']], left_on='employee_name', right_on='Payroll Name', how='left')
output_df.drop(columns=['Payroll Name'], inplace=True)
output_df.head(15)
```

```python
active_employee_file = 'Forvis Mazars Data Request/Active Employee List as of 01-07-2026.xlsx'
df_active = pd.read_excel(active_employee_file)

output_df = output_df.merge(
    df_active[["Payroll Name", "Pay Grade Code"]],
    left_on="employee_name",
    right_on="Payroll Name",
    how="left"
)
output_df.drop(columns=["Payroll Name"], inplace=True)
output_df.head(15)
```

```python
#Sanity Check
num_rows = len(output_df)
missing_pay_grade = output_df['Pay Grade Code'].isna().sum()
print(f"Total rows in output_df: {num_rows}")
print(f"Rows missing pay grade code: {missing_pay_grade}")

missing_and_not_active = output_df[output_df['Pay Grade Code'].isna() & (output_df['Status'].str.lower() != 'Active')]
print(f"Rows missing pay grade code and not active: {len(missing_and_not_active)}")
```

### Visualizations

```python
#heatmap to see correlation between Age Bracket and Tenure
ordered_age =[
    "Gen Z",
    "Millennials",
    "Gen X",
    "Baby Boomers",
    "Post War"
]

ordered_tenure = [
    "<2 years",
    "2-5 years",
    "5-10 years",
    "10-20 years",
    "20+ years"
]
output_df['age_bracket'] = pd.Categorical(output_df['age_bracket'], categories=ordered_age, ordered=True)
output_df['tenure_bracket'] = pd.Categorical(output_df['tenure_bracket'], categories=ordered_tenure, ordered=True)


crosstab = pd.crosstab(output_df['age_bracket'], output_df['tenure_bracket'])

plt.figure(figsize=(10, 6))
sns.heatmap(crosstab, annot=True, fmt='d', cmap='Blues')
plt.title('Correlation between Age Bracket and Tenure Bracket')
plt.xlabel('Tenure Bracket')
plt.ylabel('Age Bracket')
plt.tight_layout()
plt.show()
```

```python
#correlation matrix
crosstab = pd.crosstab(output_df['age_bracket'], output_df['tenure_bracket'])
#display(crosstab)

#normalize the crosstab
crosstab_percentage = crosstab.div(crosstab.sum(axis=1), axis=0) * 100

plt.figure(figsize=(10, 6))
sns.heatmap(crosstab_percentage, annot=True, fmt='.2f', cmap='Greens') 
plt.title('Percentage Correlation between Age Bracket and Tenure Bracket')
plt.xlabel('Tenure Bracket')
plt.ylabel('Age Bracket')
plt.tight_layout()
plt.show()

#chi-square test
chi2, p_value, dof, expected = chi2_contingency(crosstab)
print(f"Chi-square Statistic: {chi2}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p_value}")
```

```python
def parse_tenure(tenure_str):
    match = re.match(r"(\d+) years (\d+) months", str(tenure_str))
    if match:
        years = int(match.group(1))
        months = int(match.group(2))
        return years + months / 12
    return None

#only look at individuals with tenure of year or less and have a pay grade code
mask = output_df['tenure'].apply(parse_tenure).le(1)
one_year_or_less = output_df[mask & output_df['Pay Grade Code'].notna()]

#plot the count of pay grades
paygrade_counts = one_year_or_less['Pay Grade Code'].value_counts().sort_index()
plt.figure(figsize=(10,6))
paygrade_counts.plot(kind='bar')
plt.title('Pay Grades for Tenure  1 Year')
plt.xlabel('Pay Grade Code')
plt.ylabel('Number of Individuals')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
mask = output_df['tenure'].apply(parse_tenure).le(1)
one_year_or_less = output_df[mask & output_df['Pay Grade Code'].notna()]

pay_grade_numeric = pd.to_numeric(one_year_or_less['Pay Grade Code'], errors='coerce')
role_type = pay_grade_numeric.apply(lambda x: 'Managerial' if x >= 11 else 'Non-Managerial')

#Plot the count of managerial vs non-managerial
#1-10 is nonmanagerial, 11 and above is managerial
role_counts = role_type.value_counts().reindex(['Managerial', 'Non-Managerial'])
plt.figure(figsize=(6,5))
role_counts.plot(kind='bar', color=['#1f77b4', '#ff7f0e'])
plt.title('Managerial vs Non-Managerial (Tenure  1 Year)')
plt.xlabel('Role Type')
plt.ylabel('Number of Individuals')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
def parse_tenure_to_years(tenure_str):
    match = re.match(r"(\d+) years (\d+) months", str(tenure_str))
    if match:
        years = int(match.group(1))
        months = int(match.group(2))
        return years + months / 12
    return np.nan

has_paygrade = output_df['Pay Grade Code'].notna()
output_df_temp = output_df[has_paygrade].copy()
output_df_temp['tenure_years'] = output_df_temp['tenure'].apply(parse_tenure_to_years)

#group using pay code and output the average tenure
avg_tenure = output_df_temp.groupby('Pay Grade Code')['tenure_years'].mean().sort_index()
print('Average tenure (years) per Pay Grade Code:')
print(avg_tenure)
```
